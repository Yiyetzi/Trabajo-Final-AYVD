```python
import pandas as pd

# Carga del dataset desde el drive
df = pd.read_csv('/content/drive/MyDrive/sueño.csv')

# Vista previa
print(df.head())
print(df.columns)
```

```python
df_info = df.info()
df_head = df.head()
df_description = df.describe(include='all')

df_head, df_info, df_description
```

Se llevara a cabo el proceso de transformación del dataset

```python
#Revisión de porcentajes de valores nulos en nuestras variables

pd.set_option('display.max_rows', None)

valores_nulos = df.isnull().sum() #Suma de valores nulos
porcentaje_nulos = (df.isnull().sum()/len(df))*100 # Calculo de Porcentaje
nulos = pd.DataFrame({'Valores nulos': valores_nulos, 'Porcentaje': porcentaje_nulos})

#Imprimimos en orden descendente los datos nulos junto a sus porcentajes
nulos.sort_values('Porcentaje', ascending=False)
```

Se van a eliminar las columnas que tienen del 50% hacia arriba de valores nulos, por no tener información suficiente para analizar

```python
umbral = 0.5  # 50% de nulos como criterio
df1 = df.loc[:, df.isnull().mean() < umbral]
df1.head()
```

Se renombraran 14 variables para facilitar su comprensión

```python
#Renombramos columnas

df1 = df1.rename(columns={
    'su101': 'hrs_sue_lv',
    'su103': 'calidad_sueno',
    'su107': 'freq_toma_medic',
    'su108': 'precencia_ronquido',
    'su115': 'sleep_driving',
    'a301': 'diabetes',
    'a401': 'hipertension',
    'a502a': 'infarto',
    'a502b': 'angina_pecho',
    'a502c': 'insuficiencia_cardiaca',
    'a502d': 'enfermedad_corazon',
    'a1301': 'fuma',
    'a1312c': 'alcohol',
    'saosc': 'apnea_importante'
})
df1.head()
```

De la columna 'hrs_sue_lv', se procede a revisar las horas de sueño diarias que duerme una persona al día, cómo se encuentran valores mayores a 24 horas, se procederan a retirar ya que se consideran errores.


```python
df1['hrs_sue_lv'].value_counts()
```

```python
# se eliminaran las filas de la columna 'hrs_sue_lv' que tienen valores mayores a 24
df1 = df1[df1['hrs_sue_lv'] <= 24]
df1['hrs_sue_lv'].value_counts()
```

Su101 ahora llamada 'hrs_sue_lv', tiene un 11% de celdas vacias, por lo que se llenaran con la media a través de rangos de edad entre Jovenes (18 a 29 años), Adultos (30 a 59) y Adultos mayores (de los 60 en adelante). Se creara una nueva columna para determinar los rangos de edad ya que se seguiran ocupando

```python
#creación columna
import numpy as np

condiciones = [
    df1['edad'] <= 29,
    (df1['edad'] >= 30) & (df1['edad'] <= 59),
    df1['edad'] >= 60
]

valores = ['Joven', 'Adulto', 'Adulto Mayor']

df1['rango_edad'] = np.select(condiciones, valores,default='Desconocido')
df1.head()
```

```python
#calculo de la media por rango de edad
media_sueño_por_rango = df1.groupby('rango_edad')['hrs_sue_lv'].mean().round(2)
print(media_sueño_por_rango)
```

```python
#transformación para rellenar los campos vacios de hr de sueño con la media
df1['hrs_sue_lv'] = df1.groupby('rango_edad')['hrs_sue_lv'].transform(lambda x: x.fillna(x.mean()))
```

Procederemos a llenar el campo su108 ahora nombrado Ronca con una correlación la variable ronca. Se decidio usarse así ya que ronca es una variable que tiene todos los campos completos con los valores 0 que significa que no ronca y 1 que determina que si ronca, pero la variable su108 a demás de las opciones anteriores, también contiene la opción de "No lo sé", que se me hace una respuesta más real

```python
print(df1.columns)
```

```python
df1.loc[df1['precencia_ronquido'].isna() & (df1['ronca'] == 0), 'precencia_ronquido'] = 1
df1.loc[df1['precencia_ronquido'].isna() & (df1['ronca'] == 1), 'precencia_ronquido'] = 2
```

Se procedera a eliminar las filas que no contienen valores en los campos de diabetes ya que se encontro que no se cuenta información en las mismas filas (63) de otros padecimientos por lo que no se puede ver el impacto del sueño con las enfermedades

```python
#visualización de valores nulos
df1['diabetes'].isna().sum()
```

```python
#eliminamos filas
df1 = df1[df1['diabetes'].notna()]
df1['diabetes'].isna().sum()
```

Para la variable imc_valid_clas que califica si el paciente tiene un peso adecuado, sobrepeso y obesidad, dicha variable tiene el 3.41% de datos nulos, por lo que se procederá a llenar los valores de tres formas diferentes.

La primera opción será correlacionando la variable IMC ya que en el catalogo nos marca que si el IMC es menor a 25 imc_valid_clas es igual a 1, si el IMC es mayor a 25 y menor o igual a 30  es igual a 2 y por último si el IMC es mayor a 30 es iguial a 3.

Pero esto no nos soluciona la cantidad de valores nulos de dicha variable por lo que la segunda opción a tomar es la columna cinturaprom, pude apreciar que esta columna tiene valores disparados entre las clasificaciones imc_valid_clas, por lo que se tomaran las medianas para poder designar el valor máximo a comparar.

Por último para las celdas que sigan vacías, se tomara en cuenta el rango de edad y el sexo para darles una calificación

```python
#Correlación de acuerdo al IMC
df1.loc[df1['imc_valid_clas'].isna() & (df1['imc'] < 25), 'imc_valid_clas'] = 1
df1.loc[df1['imc_valid_clas'].isna() & (df1['imc'] > 25) & (df1['imc'] <= 30), 'imc_valid_clas'] = 2
df1.loc[df1['imc_valid_clas'].isna() & (df1['imc'] > 30), 'imc_valid_clas'] = 3
```

```python
#calculo de la media de la variable cinturaprom
media_cintura = df1.groupby('imc_valid_clas')['cinturaprom'].mean()
print(media_cintura)
```

```python
mediana_cintura = df.groupby('imc_valid_clas')['cinturaprom'].median()
print(mediana_cintura)
```

```python
#De acuerdo a las mediciones anteriores, se determino que si cinturaprom es menor o igual a 83 se agregara el valor de 1
#Si es mayor de 83 y menor o igual a 94 se colocará 2
#y mayores de 94 se colocará 3

df1.loc[df1['imc_valid_clas'].isna() & (df1['cinturaprom'] <= 83), 'imc_valid_clas'] = 1
df1.loc[df1['imc_valid_clas'].isna() & (df1['cinturaprom'] > 83) & (df1['cinturaprom'] <= 94), 'imc_valid_clas'] = 2
df1.loc[df1['imc_valid_clas'].isna() & (df1['cinturaprom'] > 94), 'imc_valid_clas'] = 3
```

```python
#visualización de valores nulos
df1['imc_valid_clas'].isna().sum()
```

```python
#calculo de la media de acuerdo al rango de edad y el sexo del paciente
media_imc = df1.groupby(['rango_edad', 'sexo'])['imc'].mean()
print(media_imc)
```

```python
#Como se puede observar la media en todos los casos esta por arriba de 25 y debajo de 30 por lo que se procederá a calificarlos con 2 al restante de celdas nulas
df1['imc_valid_clas'] = df1['imc_valid_clas'].fillna(2)
df1['imc_valid_clas'].isna().sum()
```

Una vez que ya tenemos completa la información en las columnas deseadas, se
eliminaran 33 variables de columnas que no son relevantes para el análisis de la información por redundancia en la información.

```python
#Eliminaremos otras columnas que no consideraremos necesarias para nuestro análisis

columnas_a_eliminar = [
    'folio', 'inte', 'idinsert', 'maq', 'intsel', 'intsel2', 'su102', 'su104',
    'tiempo_gen', 'est_var', 'code_upm', 'c1', 'ponde_f', 'cintura', 'cintura2',
    'rcintura', 'grupoednut', 'imc_valid', 'cat1c', 'saos', 'roncafuerte',
    'roncahab', 'roncamol', 'catberlin1', 'noreparador', 'catberlin2',
    'catberlin3', 'catberlin1a', 'catberlin2a', 'berlinscore', 'berlinriesgo',
    'berlinscoresinbmi', 'berlinriesgosinbmi', 'malacalsueno', 'impactoinsom',
    'insomhabi', 'horasueno1', 'horasueno2', 'horasueno3', 'cat2', 'hta', 'sexofem',
    'cat2c','hta','sexofem','ronca','horasueno','calidad','insom3sem'
]

# Get the list of columns that actually exist in df1
existing_columns_to_drop = [col for col in columnas_a_eliminar if col in df1.columns]

df1 = df1.drop(columns=existing_columns_to_drop)
df1.head()
```
