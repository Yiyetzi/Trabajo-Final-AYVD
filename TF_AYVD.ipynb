Se van a eliminar las columnas que tienen del 50% hacia arriba de valores nulos, por no tener información suficiente para analizar

Se renombraran 14 variables para facilitar su comprensión

```python
#Renombramos columnas

df1 = df1.rename(columns={
    'su101': 'hrs_sue_lv',
    'su103': 'calidad_sueno',
    'su107': 'freq_toma_medic',
    'su108': 'precencia_ronquido',
    'su115': 'sleep_driving',
    'a301': 'diabetes',
    'a401': 'hipertension',
    'a502a': 'infarto',
    'a502b': 'angina_pecho',
    'a502c': 'insuficiencia_cardiaca',
    'a502d': 'enfermedad_corazon',
    'a1301': 'fuma',
    'a1312c': 'alcohol',
    'saosc': 'apnea_importante'
})
df1.head()
```

De la columna 'hrs_sue_lv', se procede a revisar las horas de sueño promedio diarias que duerme una persona al día de lunes a viernes.
Se encuentraron valores mayores a 24 horas, se procederan a retirar ya que se consideran errores.


```python
df1['hrs_sue_lv'].value_counts()
```

```python
# se eliminaran las filas de la columna 'hrs_sue_lv' que tienen valores mayores a 24
df1 = df1[df1['hrs_sue_lv'] <= 24]
df1['hrs_sue_lv'].value_counts()
```

Su101 ahora llamada 'hrs_sue_lv', tiene un 11% de celdas vacias, por lo que se llenaran con la media a través de rangos de edad entre Jovenes (18 a 29 años), Adultos (30 a 59) y Adultos mayores (de los 60 en adelante). Se creara una nueva columna para determinar los rangos de edad ya que se seguiran ocupando

```python
#creación columna
import numpy as np

condiciones = [
    df1['edad'] <= 29,
    (df1['edad'] >= 30) & (df1['edad'] <= 59),
    df1['edad'] >= 60
]

valores = ['Joven', 'Adulto', 'Adulto Mayor']

df1['rango_edad'] = np.select(condiciones, valores,default='Desconocido')
df1.head()
```

```python
#calculo de la media por rango de edad
media_sueño_por_rango = df1.groupby('rango_edad')['hrs_sue_lv'].mean().round(2)
print(media_sueño_por_rango)
```

```python
#transformación para rellenar los campos vacios de hr de sueño con la media
df1['hrs_sue_lv'] = df1.groupby('rango_edad')['hrs_sue_lv'].transform(lambda x: x.fillna(x.mean()))
```

Procederemos a llenar el campo su108 ahora nombrado Ronca con una correlación la variable ronca. Se decidio usarse así ya que ronca es una variable que tiene todos los campos completos con los valores 0 que significa que no ronca y 1 que determina que si ronca, pero la variable su108 a demás de las opciones anteriores, también contiene la opción de "No lo sé", que se me hace una respuesta más real

Se procedera a eliminar las filas que no contienen valores en los campos de diabetes ya que se encontro que no se cuenta información en las mismas filas (63) de otros padecimientos por lo que no se puede ver el impacto del sueño con las enfermedades

Para la variable imc_valid_clas que califica si el paciente tiene un peso adecuado, sobrepeso y obesidad, dicha variable tiene el 3.41% de datos nulos, por lo que se procederá a llenar los valores de tres formas diferentes.

La primera opción será correlacionando la variable IMC ya que en el catalogo nos marca que si el IMC es menor a 25 imc_valid_clas es igual a 1, si el IMC es mayor a 25 y menor o igual a 30  es igual a 2 y por último si el IMC es mayor a 30 es iguial a 3.

Pero esto no nos soluciona la cantidad de valores nulos de dicha variable por lo que la segunda opción a tomar es la columna cinturaprom, pude apreciar que esta columna tiene valores disparados entre las clasificaciones imc_valid_clas, por lo que se tomaran las medianas para poder designar el valor máximo a comparar.

Por último para las celdas que sigan vacías, se tomara en cuenta el rango de edad y el sexo para darles una calificación

```python
#Correlación de acuerdo al IMC
df1.loc[df1['imc_valid_clas'].isna() & (df1['imc'] < 25), 'imc_valid_clas'] = 1
df1.loc[df1['imc_valid_clas'].isna() & (df1['imc'] > 25) & (df1['imc'] <= 30), 'imc_valid_clas'] = 2
df1.loc[df1['imc_valid_clas'].isna() & (df1['imc'] > 30), 'imc_valid_clas'] = 3
```

```python
#calculo de la media de la variable cinturaprom
media_cintura = df1.groupby('imc_valid_clas')['cinturaprom'].mean()
print(media_cintura)
```

```python
mediana_cintura = df.groupby('imc_valid_clas')['cinturaprom'].median()
print(mediana_cintura)
```

```python
#De acuerdo a las mediciones anteriores, se determino que si cinturaprom es menor o igual a 83 se agregara el valor de 1
#Si es mayor de 83 y menor o igual a 94 se colocará 2
#y mayores de 94 se colocará 3

df1.loc[df1['imc_valid_clas'].isna() & (df1['cinturaprom'] <= 83), 'imc_valid_clas'] = 1
df1.loc[df1['imc_valid_clas'].isna() & (df1['cinturaprom'] > 83) & (df1['cinturaprom'] <= 94), 'imc_valid_clas'] = 2
df1.loc[df1['imc_valid_clas'].isna() & (df1['cinturaprom'] > 94), 'imc_valid_clas'] = 3
```

```python
#visualización de valores nulos
df1['imc_valid_clas'].isna().sum()
```

```python
#calculo de la media de acuerdo al rango de edad y el sexo del paciente
media_imc = df1.groupby(['rango_edad', 'sexo'])['imc'].mean()
print(media_imc)
```

```python
#Como se puede observar la media en todos los casos esta por arriba de 25 y debajo de 30 por lo que se procederá a calificarlos con 2 al restante de celdas nulas
df1['imc_valid_clas'] = df1['imc_valid_clas'].fillna(2)
df1['imc_valid_clas'].isna().sum()
```

## **Parte 3:**

Se revisará de nuestras columnas cuales son de tipo categorico para convertirlas a númerico para poder realizar la correlación. Posteriormente convertiremos estas variables a númericas.

```python
# Detectar columnas tipo object
columnas_object = df1.select_dtypes(include='object').columns
print(columnas_object)

# Convertir a tipo numérico
for col in columnas_object:
    df1[col] = df1[col].astype('category')
    print(f"\nColumna: {col}")
    print("Categorías detectadas:", df1[col].cat.categories.tolist())
    df1[col] = df1[col].cat.codes  # Asignar códigos numéricos

print("\nTipos de datos después de conversión:")
print(df1.dtypes)
```

```python
#Creamos nueva columna 'tiene_enfermedad' para hacer correlación de acuerdo a si tiene diabetes, hipertensión o cualquiera de las enfermedades en el dataset
columnas_enfermedades = [
    'diabetes',
    'hipertension',
    'infarto',
    'angina_pecho',
    'insuficiencia_cardiaca',
    'enfermedad_corazon'
]

df1['tiene_enfermedad'] = (df1[columnas_enfermedades] == 1).any(axis=1).astype(int)
df1.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Definir las columnas de interés
columnas_interes = ['calidad_sueno', 'sexo', 'tiene_enfermedad',
                    'imc_valid_clas', 'insomnio', 'rango_edad']
df_subset = df1[columnas_interes]

# Calcular la matriz de correlación de Pearson
correlacion = df_subset.corr(method='pearson')

# Mostrar la matriz
print("Matriz de correlación entre las variables seleccionadas:")
print(correlacion)

# Graficar la matriz con mapa de calor
plt.figure(figsize=(8, 6))
sns.heatmap(correlacion, annot=True, cmap='coolwarm', fmt='.2f', square=True)
plt.title('Matriz de correlación (Pearson)')
plt.tight_layout()
plt.show()
```

De acuerdo a la correlación observada entre las variables seleccionadas, se puede apreciar que el insomnio es el que tiene una mayor correlación positiva con la calidad de sueño, por debajo quedando si tiene una enfermedad y el IMC.

En cuestión de la vatiable tiene enfermedad la correlacíon más alta es respecto al IMC, seguido de si tiene insomnio, esto nos dice que el indice de masa corporal y el insomnio, si tiene correlación positiva, aunque es débil.

Por último se de acuerdo imc_valid_clas vs rango_edad, las personas más jóvenes tienden a estar en categorías más saludables de IMC, pero la relación es leve.


```python
# Calcular correlación de Spearman para capturar relaciones monótonas no lineales.
correlacion_spearman = df_subset.corr(method='spearman')

# Mostrar la matriz
print("Matriz de correlación (Spearman):")
print(correlacion_spearman)

# Mapa de calor
plt.figure(figsize=(8, 6))
sns.heatmap(correlacion_spearman, annot=True, cmap='coolwarm', fmt='.2f', square=True)
plt.title('Matriz de correlación (Spearman)')
plt.tight_layout()
plt.show()
```

Aplicamos la correlación de Spearman para ver si hay relación no lineal entre los valores, pero como podemos observar los valores no difieren mucho con los arrojados por la correlación de Pearson.

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

variables = ['sexo', 'tiene_enfermedad', 'imc_valid_clas', 'insomnio', 'rango_edad']

# Estandarizar (muy importante en PCA)
X = StandardScaler().fit_transform(df1[variables])

# Aplicar PCA
pca = PCA(n_components=2)
componentes = pca.fit_transform(X)

# Crear DataFrame con los resultados
df_pca = pd.DataFrame(data=componentes, columns=['PC1', 'PC2'])
df_pca['calidad_sueno'] = df1['calidad_sueno']  # opcional para colorear por calidad

# Visualizar
plt.figure(figsize=(8,6))
sns.scatterplot(data=df_pca, x='PC1', y='PC2', hue='calidad_sueno', palette='coolwarm')
plt.title('PCA de variables relacionadas al sueño')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.legend(title='Calidad Sueño')
plt.grid(True)
plt.tight_layout()
plt.show()

# Ver la varianza explicada
print("Varianza explicada por cada componente:")
print(pca.explained_variance_ratio_)
```

Cómo nuestros puntos no tienen una agrupación, veremos que variables influyen más en el principal componente del PCA

```python
# Obtener los pesos (loadings)
loadings = pd.DataFrame(
    pca.components_.T,
    columns=['PC1', 'PC2'],
    index=variables
)
print(loadings)
```

Para el PC1 podemos ver:
- imc_valid_clas tiene el peso más alto (0.598), por lo tanto, el índice de masa corporal es la variable que más influye en PC1.
- También influyen positivamente tiene_enfermedad (0.458) e insomnio (0.329).

En el caso del PC2:
- Las variables que más aportan son insomnio (0.623) y rango_edad (0.603), ambas de forma positiva.
